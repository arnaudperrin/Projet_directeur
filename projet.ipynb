```python
coef_exp = np.exp(results.params)
coef_exp = pd.DataFrame(coef_exp)
coef_exp.columns = ['OR']
coef_exp
intervalle_exp = np.exp(results.conf_int())
intervalle_exp.columns = [['lower','upper']]
intervalle_exp
coef_exp = pd.merge(coef_exp, intervalle_exp, left_index=True, right_index=True)
p_values = results.pvalues
p_values = pd.DataFrame(p_values)
p_values.columns = ['p-value']
coef_exp = pd.merge(coef_exp, p_values, left_index=True, right_index=True)
coef_exp['p-value'] = coef_exp['p-value'].round(4)
coef_exp.columns = [['OR','lower','upper','p-value']]
coef_exp = coef_exp.sort_values(by=('OR',), ascending=False)
coef_exp = coef_exp.drop('Intercept')
coef_exp.index = [index.split('[')[-1].replace('T.', '').replace(']', '') for index in coef_exp.index]
coef_exp
```
