```python


echantillon['Directeur de these'] = echantillon['Directeur de these'].astype(str)

for i in range(7):
    echantillon[f'Directeur de these {i + 1}'] = ''

def decomposer_directeurs(row):
    directeurs = row['Directeur de these'].split(',') 
    for i, directeur in enumerate(directeurs):
        echantillon.at[row.name, f'Directeur de these {i + 1}'] = directeur


echantillon.apply(decomposer_directeurs, axis=1)

echantillon
```

```python
echantillon["etu_dir"] = 0
nb_lignes,nb_columns = echantillon.shape

def check_auteur_in_directeur(row):
    auteur = row["Auteur"]
    directeurs = row[2:]  # Sélectionnez toutes les colonnes à partir de la deuxième (Directeur de these 1, Directeur de these 2, ...)
    return 1 if auteur in directeurs.values else 0

# Appliquez la fonction à chaque ligne du DataFrame et mettez à jour la colonne "etu_dir"
echantillon["etu_dir"] = echantillon.apply(check_auteur_in_directeur, axis=1)

echantillon["etu_dir"].value_counts()
```
